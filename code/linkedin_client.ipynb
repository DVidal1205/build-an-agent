```python
import logging
logging.basicConfig(level=logging.INFO)

from IPython.display import Markdown, Image, display
from pathlib import Path

from linkedin_agent import async_create_linkedin_post
```

```python
## Configure Your Post
```

```python
# Your content and settings
prompt = "Write my goodbye post for my internship at NVIDIA."
image_filename = "hike.png"  # Image should be in images/ folder

# Optional: Style preferences (1-5 scale)
grammar_level = 5      # 1=terrible, 5=proper
emoji_level = 3        # 1=minimal, 5=maximum  
hashtag_level = 3      # 1=few, 5=many
ragebait_level = 3     # 1=humble, 5=egotistical
inspirational_level = 5 # 1=realistic, 5=dream-big
informational_level = 3 # 1=opinion, 5=data-rich
```

```python
## Image Preview
```

```python
# Display the image
image_path = f"images/{image_filename}"
if Path(image_path).exists():
    print(f"ðŸ“¸ Using image: {image_path}")
    display(Image(image_path, width=400))
else:
    print(f"âš ï¸  Image not found: {image_path}")
    print("Will generate text-only post")
    image_path = None
```

```python
## Generate LinkedIn Post
```

```python
# Generate the LinkedIn post
print("ðŸš€ Generating LinkedIn post...")
print(f"ðŸ“ Prompt: {prompt}")
print(f"ðŸŽ¨ Style: Grammar={grammar_level}, Emojis={emoji_level}, Hashtags={hashtag_level}")
print("-" * 80)

result = await async_create_linkedin_post(
    initial_prompt=prompt,
    image_path=image_path,
    post_type="general",
    grammar_level=grammar_level,
    emoji_level=emoji_level,
    hashtag_level=hashtag_level,
    ragebait_level=ragebait_level,
    inspirational_level=inspirational_level,
    informational_level=informational_level
)

print("\nðŸŽ¯ LinkedIn post generated!")
```

```python
## Final Result
```

```python
# Display the generated post
if result and result.get("final_post"):
    display(Markdown(result["final_post"]))
else:
    print("âŒ Failed to generate post")
    if result:
        print(f"Available keys: {list(result.keys())}")
```
